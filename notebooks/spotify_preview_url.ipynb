```python
import requests
import base64
```

```python
#authentification
client_id = os.environ['SPOTIFY_CLIENT_ID']
client_secret = os.environ['SPOTIFY_CLIENT_SECRET']

# Encode client ID and client secret
client_credentials = f"{client_id}:{client_secret}"
base64_credentials = base64.b64encode(client_credentials.encode()).decode()

# Request access token
auth_url = "https://accounts.spotify.com/api/token"
headers = {"Authorization": f"Basic {base64_credentials}"}
data = {"grant_type": "client_credentials"}
response = requests.post(auth_url, headers=headers, data=data)
response_data = response.json()
access_token = response_data["access_token"]
```

```python
track_id = "5SuOikwiRyPMVoIQDJUgSV"

# API request
api_url = f"https://api.spotify.com/v1/tracks/{track_id}"
headers = {"Authorization": f"Bearer {access_token}"}
response = requests.get(api_url, headers=headers)
response_data = response.json()

# Extract the preview URL from the response
preview_url = response_data["preview_url"]
```

```python
def get_preview(track_id):
    api_url = f"https://api.spotify.com/v1/tracks/{track_id}"
    headers = {"Authorization": f"Bearer {access_token}"}
    response = requests.get(api_url, headers=headers)
    response_data = response.json()
    
    # Extract the preview URL from the response
    return response_data["preview_url"]
```
