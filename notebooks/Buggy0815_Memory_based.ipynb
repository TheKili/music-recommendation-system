```python
import pandas as pd
import os
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics.pairwise import cosine_similarity
import sys
import os
```

```python
# Get the parent directory path
parent_dir = os.path.dirname(os.path.abspath(''))
# Add the parent directory to sys.path
sys.path.append(parent_dir)
```

```python
data_path = os.path.join(os.path.dirname(os.getcwd()), 'raw_data/dataset.csv')
```

```python
df = pd.read_csv(data_path, index_col = 0)
```

```python
df.head()
```

```python
df['track_genre'].value_counts()
```

```python
df.loc[df.duplicated('track_id', keep = False)].sort_values('track_id')
```

```python
genres = df.groupby('track_id').agg({'track_genre' : list})
```

```python
genres['track_genre'].apply(lambda x: len(x)).value_counts()
```

```python
df = df.drop_duplicates(subset = 'track_id')
```

```python
df = df.drop(columns = 'track_genre')
```

```python
df = pd.merge(left = df, right = genres, on= 'track_id')
```

# Ab hier Paul

```python
audio_feats = [
     'popularity',
     'duration_ms',
     'explicit',
     'danceability',
     'energy',
     'key',
     'loudness',
     'mode',
     'speechiness',
     'acousticness',
     'instrumentalness',
     'liveness',
     'valence',
     'tempo'
]
```

```python
scaler = MinMaxScaler()
```

```python
df_audio = df.loc[:, audio_feats]
```

```python
df_audio_scaled = pd.DataFrame(scaler.fit_transform(df_audio), columns = df_audio.columns)
```

```python
df_audio_scaled = df_audio_scaled.set_index(df['track_id'])
```

ab hier habe ich die Ähnlichkeit mit cosine_similarity berechnet. Ab hier kannst du andere Ähnlichkeitsmetriken ausprobieren

```python
similarities = cosine_similarity(X = df_audio_scaled.iloc[[0]], Y = df_audio_scaled).T[:,0]
```

```python
df_sim = pd.DataFrame({'track_id' : df['track_id'],
              'similarity' : similarities,
              'track_name' : df['track_name'],
              'artist' : df['artists']})
```

```python
df_sim.sort_values('similarity', ascending = False)[1:5]
```

```python
df_sim.sort_values('similarity', ascending = False)
```

```python
pd.DataFrame({'tst' : cosine_similarity(Y = df_audio_scaled.iloc[[0]], X = df_audio_scaled),
              'title' : df['track_name']}, index = df_audio_scaled.index)
```
